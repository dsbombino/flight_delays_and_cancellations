#### There are several columns with nan values.
For the purpose of this study we will set some assumptions in order to handle this missing data:
- Nan values in Delay's columns will be considered as cero delay. 
    - All nan values in Delay's columns will be replaced by cero
- Rows having nan values in Departure and Arrival Time columns are considered irrelevant for the purpose of this study.
    - All rows having nan values in Departure and Arrival Time columns will be deleted.

```python
#Replace nan values in delay columns by cero
delay_columns = ['DEPARTURE_DELAY',
                 'ARRIVAL_DELAY', 
                 'AIR_SYSTEM_DELAY', 
                 'SECURITY_DELAY', 
                 'AIRLINE_DELAY', 
                 'LATE_AIRCRAFT_DELAY', 
                 'WEATHER_DELAY']
delay_columns_values = {col:0 for col in delay_columns}
flights.fillna(value=delay_columns_values, inplace=True)
```

#### Now that both df were merged and missing data was handled, lets perform some other transformation in our data:
- Create a "DATE" column by combining the information of YEAR, MONTH and DAY.
- Create a new column with the name of the day (MONDAY, TUESDAY, etc.)
- Converting the columns containing time information to a more suitable format.

```python
#Create a "DATE" column by combining the information of YEAR, MONTH and DAY.

#Create a pandas series with the date information by concatenating YEAR, MONTH and DAY columns
flights_date= flights['YEAR'].map(str) + '-' + flights['MONTH'].map(str) + '-' + flights['DAY'].map(str)
#Insert a DATE column with the flights dates in position cero
flights.insert(0, 'DATE', flights_date)
```

```python
#Change the dtype of DATE column to datetime64
flights['DATE'] = flights['DATE'].astype('datetime64')
#Get a sample of the data
flights[['DATE', 'YEAR', 'MONTH', 'DAY']].sample(10)
```

As it can be seen, a DATE column was insert on position cero, containing the date with the format yyyy-mm-dd

#### In the next step, I will change the dtype of the columns containing time information from string to datetime formats

```python
#Converting the columns containing time information from string to hours.

#Create a list with all the columns with time information
time_columns = ['SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'WHEELS_OFF', 'WHEELS_ON', 'SCHEDULED_ARRIVAL', 'ARRIVAL_TIME']
```

```python
#All the time values are string formed by 4 digits, where each digits represent hhmm
#Create a function to insert ":" between hh and mm
def insert(input_str, position, str_to_insert):
    return ''.join((input_str[:position], str_to_insert, input_str[position:]))
```

```python
#Apply the insert function to our time columns
for col in time_columns:
    flights[col] = flights[col].apply(insert, args=(2, ':'))
```

```python
#Get a sample of our data
print(display(flights[time_columns].sample(5)))
#Check the format of our values
print(flights.loc[1, 'SCHEDULED_DEPARTURE'], type(flights.loc[1, 'SCHEDULED_DEPARTURE']))
```

Now, we see that the two points were correctly insert between hh and mm, however, the values are still string.
#### Let’s change the format from string to datetime.

```python
#Convert our values from strings to datetime
for col in time_columns:
    flights[col].replace({'24:00':'00:00'}, inplace=True)
    flights[col] = flights[col].apply(lambda x: datetime.strptime(x, '%H:%M').time())
```

```python
#Get a sample of our data
print(display(flights[time_columns].sample(5)))
#Check the format of our values
print(flights.loc[1, 'SCHEDULED_DEPARTURE'], type(flights.loc[1, 'SCHEDULED_DEPARTURE']))
```

Now we have converted all the values in time columns to datetime types.

#### Let’s figure out the total airtime and travel distance per company and the total for all of those companies.

```python
#Get the total airtime and travel distance (the data is in minutes)
total_airtime_min = flights.AIR_TIME.sum()
total_airtime_hrs = total_airtime_min//60
total_airtime_days = total_airtime_hrs//24
total_airtime_years = total_airtime_days//365
print('The total airtime for all companies was around: ', 
      '{} minutes or,'.format(total_airtime_min),
      '{} hours or,'.format(total_airtime_hrs),
      '{} days or,'.format(total_airtime_days),
      '{} years'.format(total_airtime_years),
     sep='\n')
```

```python
#Get the total traveled distance (the data is in miles)
total_distance_miles = flights.DISTANCE.sum()
total_distance_kms = total_distance_miles*1.61
hearth_perimeter = 40075 #In Kms
print('The total distance for all companies was around: ', 
      '{} miles or,'.format(total_distance_miles),
      '{} kms or,'.format(total_distance_kms),
      '{} trips around the Earth'.format(total_distance_kms//hearth_perimeter),
     sep='\n')
```

That is a lot of time in the air and trips around the Globe!!

```python
#Calculate average speed (kms/hr)
avg_speed = total_distance_kms/total_airtime_hrs
print('The average speed for all companies was: {} km/hr'.format(avg_speed))
```

```python
#### Now let's calculate the total airtime and traveled distance per company
airtime_distance_per_company = flights.groupby('AIRLINE_NAME').agg({'AIR_TIME':'sum',
                                                                   'DISTANCE':'sum'})
airtime_distance_per_company
```

#### Let’s plot this information using a pie chart

```python
#Plot air time information
airtime_distance_per_company.AIR_TIME.plot(kind='pie', figsize=(20, 10))
plt.pyplot.title('Percentage of air time per company')
```

In this case, it can be seen that three companies that represent the biggest part of the pie. Delta Air Lines Inc., American Airlnes Inc., and Southwest Airlines Co.
#### We can check this by calculating the cumulative sum of the AIR_TIME per company.

```python
#Insert columns with the percentage per company for both, AIR_TIME and DISTANCE columns
airtime_distance_per_company['AIR_TIME_%'] = round(100*(airtime_distance_per_company['AIR_TIME']/airtime_distance_per_company['AIR_TIME'].sum()), 2)
airtime_distance_per_company['DISTANCE_%'] = round(100*(airtime_distance_per_company['DISTANCE']/airtime_distance_per_company['DISTANCE'].sum()), 2)

#Create two different df for each metric and calculate the cumulative percentage per company

#Air Time df
airtime_per_company = airtime_distance_per_company['AIR_TIME_%'].sort_values(ascending = False).to_frame()
airtime_per_company['AIR_TIME_%_CUMSUM'] = airtime_per_company['AIR_TIME_%'].cumsum()
print(display(airtime_per_company))
```

By checking the cumulative sum, we can corroborate the information from the graph, four companies totalize more than the 60% of the total air time.

```python
#Plot distance information
airtime_distance_per_company.DISTANCE.plot(kind='pie', figsize=(20, 10))
plt.pyplot.title('Percentage of traveled distance per company')
```

For the distance information, we see exactly the same trend from the air time graph.

### Number of departure and arrival delays per Airline company.

```python
#number of departure and arrival delays per Airline company.
departure_delay_bins = [flights.DEPARTURE_DELAY.min(), 0, flights.DEPARTURE_DELAY.max()]
departure_bin_names = ['NOT_A_DELAY', 'A_DELAY']
#Insert the DEPARTURE_DELAY_INFO column
flights['DEPARTURE_DELAY_INFO'] = pd.cut(flights['DEPARTURE_DELAY'], departure_delay_bins, labels=departure_bin_names, include_lowest=True)

#number of arrival delays per Airline company
arrival_delay_bins = [flights.ARRIVAL_DELAY.min(), 0, flights.ARRIVAL_DELAY.max()]
arrival_bin_names = ['NOT_A_DELAY', 'A_DELAY']
#Insert the ARRIVAL_DELAY_INFO columns
flights['ARRIVAL_DELAY_INFO'] = pd.cut(flights['ARRIVAL_DELAY'], arrival_delay_bins, labels=arrival_bin_names, include_lowest=True)
```

```python
#Create a df with the informarmation of the number of delays per company
delays_per_company = flights.groupby('AIRLINE_NAME').agg({'DEPARTURE_DELAY_INFO':'value_counts',
                                                                   'ARRIVAL_DELAY_INFO':'value_counts'})
delays_per_company
```

```python
#Bar graphs of the number of delays per airiline company
colors=['orangered', 'dodgerblue', 'orangered', 'dodgerblue']
delays_per_company.unstack().plot(kind='bar', figsize=(20,10), color=colors)
plt.pyplot.xlabel('AIRLINE NAMES')
plt.pyplot.ylabel('NUMBER OF FLIGHTS')
plt.pyplot.title('RELATIONS BETWEEN DEPARTURE AND ARRIVAL DELAYS')
```

As it can be seen in the picture, for most of the companies the number of flights with delays (red color) were lower than the number of flights without delays. However, the ratio of the number of flights with delays vs flights with no delays, seems to be higher than 0.5 (or at least pretty close) for every company, which indicates that for every three flights, at least one will be delayed (3 flights --> 1-delayed, 2-no delayed)

### Calculate delay ratios per company (delay/no delay)

```python
#Add a column with the sum of deperture and arrivale delays and no delays per company to be used to compute de delay ratio per company
delays_per_company['TOTAL_DELAY_INFORMATION'] = delays_per_company.DEPARTURE_DELAY_INFO + delays_per_company.ARRIVAL_DELAY_INFO
delays_per_company.head(6)
```

```python
#Create a dictionary with the ratio per company
delay_ratios_dic={}
for company in delays_per_company.index.levels[0]:
        delay_ratio = delays_per_company.loc[(company, 'A_DELAY'), 'TOTAL_DELAY_INFORMATION'] / delays_per_company.loc[(company, 'NOT_A_DELAY'), 'TOTAL_DELAY_INFORMATION']
        delay_ratios_dic.update({company:round(delay_ratio, 2)})
delay_ratios_dic = sorted(delay_ratios_dic.items(), key=lambda x: x[1])
delay_ratios_dic
```

As seen in the picture "RELATIONS BETWEEN DEPARTURE AND ARRIVAL DELAYS", the relation between the number of delays and no delays is close to 0.5 or higher for every airline company. The airlines with the better ratios are Alaska Airlines Inc., Delta Air Lines Inc., and Hawaiian Airlines Inc. with a little bit more of the 50% of their flights with no delays. On the other hands, the companies with the worst ratios are Southwest Airlines Co., Frontier Airlines Inc., United Air Lines Inc., and Spirit Air Lines with more than the 70% of their flights with delays.

Going back to the image "RELATIONS BETWEEN DEPARTURE AND ARRIVAL DELAYS", another conclusion that could be made, is that there is a strong relation between the number of arrival delays and departure delays (which usually makes sense). This information could be contrasted by computating the correlation between these two variables.
